```python
import geopandas as gpd
import folium
import pandas as pd
```

```python
m2_amplitude = []
for year in range(1999, 2016):
    file = 'm2_amplitude_shp_100m_' + str(year) + '/' + 'm2_amplitude_shp_100m_' + str(year) + '.shp'
    m2_amplitude.append(gpd.read_file(file))
```

```python
m2_phase = []
for year in range(1997, 2016):
    file = 'm2_phase_shp_100m_' + str(year) + '/' + 'm2_phase_shp_100m_' + str(year) + '.shp'
    m2_phase.append(gpd.read_file(file))
```

```python
salinity_01 = []
for year in range(1996, 2014):
    file = 'salinity_01q_shp_100m_' + str(year) + '/' + 'salinity_01q_shp_100m_' + str(year) + '.shp'
    salinity_01.append(gpd.read_file(file))
```

```python
salinity_99 = []
for year in range(1996, 2001):
    file = 'salinity_99q_shp_100m_' + str(year) + '/' + 'salinity_99q_shp_100m_' + str(year) + '.shp'
    salinity_99.append(gpd.read_file(file))
for year in range(2002, 2014):
    file = 'salinity_99q_shp_100m_' + str(year) + '/' + 'salinity_99q_shp_100m_' + str(year) + '.shp'
    salinity_99.append(gpd.read_file(file))
```

```python
def extractHigherValue(row):
    return float(row.split(' ')[4])

def extractHigherValue2(row):
    return float(row.split(' ')[2])

for dataset in m2_amplitude:
    dataset['Amplitude'] = dataset.polytype.apply(lambda row: extractHigherValue(row))

for dataset in m2_phase:
    dataset['Phase'] = dataset.polytype.apply(lambda row: extractHigherValue2(row))

for dataset in salinity_01:
    dataset['Salzgehalt'] = dataset.polytype.apply(lambda row: extractHigherValue(row))

for dataset in salinity_99:
    dataset['Salzgehalt'] = dataset.polytype.apply(lambda row: extractHigherValue(row))
```

```python
m = m2_amplitude[0].explore(
    column="Amplitude",  # make choropleth based on "POP2010" column
    legend=True,  # show legend
    tooltip=False,  # hide tooltip
    popup=["polytype"],  # show popup (on-click)
    name="Amplitude 1999",  # name of the layer in the map
    vmin=0,
    vmax=2
)
for i in (range(1, len(m2_amplitude))):
    name = 'Amplitude' + str(1999+i)
    m2_amplitude[i].explore(
        m=m,  # pass the map object
        column="Amplitude",
        legend=False,
        tooltip=False,
        popup=["polytype"],
        name=name,  # name of the layer in the map
        vmin=0,
        vmax=2
    )

folium.LayerControl().add_to(m)  # use folium to add layer control

m  # show map
```
